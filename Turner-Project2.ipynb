## Inputs read from csv file

```python
# Read in the input data from HawaiiData.csv
import csv

header = []
data = []

with open('/content/gdrive/MyDrive/Colab Notebooks/Input.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        if line_count == 0:
            header = row
            line_count += 1
        else:
            row[0] = int(row[0]) # convert to an integer 
            row[1] = float(row[1])  # convert to a float
            data.append(row)
            line_count += 1
```

```python
print("Total Instances:", len(data))
print(header)
```

## Code a Linear Regression model using sklearn

```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

## Split the data into test/train

```python
from sklearn.model_selection import train_test_split

# SAT values in X
X = []
for row in data:
    X.append(row)

# Accepted to Harvard
y = [] 
for row in data:
    if row[0] > 1760:
        y.append(1)
    else: 
        y.append(0)

X_train, X_test, y_train, y_test = train_test_split(np.array(X), np.array(y).reshape(-1,1), test_size = 0.2)
```

## Fit the model to the training data

```python
lin_regressor = LinearRegression()
lin_regressor.fit(X_train, y_train)
```

## Make predictions using predict()

```python
y_predicted = lin_regressor.predict(X_test)
print(y_predicted) 
```

## Measure the performance of the model

```python
# Measure the performance speed of the model
# Code taken from https://datatofish.com/measure-time-to-run-python-script/
import time
startTime = time.time()

lin_regressor = LinearRegression()
lin_regressor.fit(X_train, y_train)
lin_regressor.predict(X_test)

executionTime = (time.time() - startTime)
print('Execution time in seconds: ' + str(executionTime))
```

## Analysis of Results

Accuracy

```python
from __future__ import print_function
# accuracy = (tp + tn) / (tp + fp + tn + fn)
tp = 0
tn = 0
fp = 0
fn = 0

rounded_y_predicted = []
for row in y_predicted:
    rounded_y_predicted.append(round(row[0]))

for i, row in enumerate(y_test):
    for element in row:
        if (rounded_y_predicted[i] == 1) & (element == 1):
            tp = tp + 1
        elif (rounded_y_predicted[i] == 0) & (element == 0):
            tn = tn + 1
        elif (rounded_y_predicted[i] == 0) & (element == 1):
            fn = fn + 1
        elif (rounded_y_predicted[i] == 1) & (element == 0):
            fp = fp + 1

accuracy = (tp + tn) / (tp + fp + tn + fn)

print("The Accuracy of this model is:", accuracy)
    
```

Confusion Matrix

```python
# confusion matrix
matrix =[[[tp],[fn]],
          [[fp], [tn]]]

for row in matrix:
    print(row)
```

F1 Score

```python
# Compute the F1 score  (F1 Score = 2 * (Precision * Recall) / (Precision + Recall))

precision = tp / (tp + fp)
recall = tp / (tp + fn)

f1_score = 2 * (precision * recall) / (precision + recall)

print(f1_score)
```
